```python



# Your data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
total_gen = [15145870, 17840440, 28059090, 34134550, 41235270, 48239290, 50559990, 44243690, 36102070, 23922630, 13946980, 12813520]
gen_below_10 = [129181, 104161, 112162, 172082, 94245, 30443, 12480, 176880, 59081, 131768, 176861, 158329]
gen_above_10 = [15016690, 17736280, 27946930, 33962470, 41141030, 48208850, 50547510, 44066810, 36042990, 23790860, 13457120, 12655190]
consumption_above_10 = [14198050, 15417540, 21654330, 24484850, 28307990, 31022190, 32357090, 29589870, 25967740, 20267630, 13229160, 12384610]
percentage_total = [93.72, 90.43, 77.14, 71.74, 68.57, 64.22, 63.93, 66.72, 71.93, 84.72, 94.85, 96.65]

# Create figure and axis objects
fig, ax1 = plt.subplots()

# Bar chart for the generation data
bar_width = 0.25
index = np.arange(len(months))
bar1 = ax1.bar(index - bar_width, gen_below_10, bar_width, label='Generation Below 10 MW', color='#000090')
bar2 = ax1.bar(index - bar_width, gen_above_10, bar_width, bottom=gen_below_10, label='Generation Above 10 MW', color='g')
bar3 = ax1.bar(index, consumption_above_10, bar_width, bottom=gen_below_10, label='Consumption Above 10 MW', color='#000000')
# Line chart for the percentage of total consumption
ax2 = ax1.twinx()
line1 = ax2.plot(index, percentage_total, label='Percentage of Total Consumption', color='r', marker='o')

# Labels, title and legend
ax1.set_xlabel('Month')
ax1.set_ylabel('Generation (log(kWh))', color='b')
ax2.set_ylabel('Percentage of Total Consumption', color='r')
ax1.set_xticks(index)
ax1.set_xticklabels(months)
# ax1.set_yscale('log')
# ax1.set_ylim(1e0, 1e8)
legend_fontsize = 4  # Smaller font size for the legend
ax1.legend(loc='upper left', fontsize=legend_fontsize, prop={'size': legend_fontsize})
ax2.legend(loc='upper right', fontsize=legend_fontsize, prop={'size': legend_fontsize})

plt.title('Monthly Energy Metrics')

# Show the plot
plt.show()
```
